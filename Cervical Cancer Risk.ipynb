```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('kag_risk_factors_cervical_cancer.csv')
pd.set_option('display.max_columns', None)
data.head()
```

```python
data.info()
```

#### MEANING OF VARIABLES

```python
data = data.apply(pd.to_numeric, errors='coerce')
data = data.replace('?', np.nan)
nan_counts = data.isnull().sum()
nan_counts
```

```python
data.describe(include= 'all')
```

```python
# We will throw out two columns that have over 90% Nan values
data.drop(columns=['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'], inplace=True)
data.head()
```

```python
# Replacing NaN values with mean
data = data.fillna(data.mean())
```

```python
# correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
corr = sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.show()
```

```python
cancer_correlations = correlation_matrix['Biopsy']
print(cancer_correlations)
```

From the correlation table, we see that the highest correlated with the diagnosis of cervical cancer is positive Schiller test, positive Hinselmann test has a slightly lower correlation, while the other correlation values ​​are weak

# Prediction of biopsy result

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler


target = data['Biopsy']
features = data.drop(columns=['Biopsy'])           

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

data['Biopsy'].value_counts()
```


As we can see, the target variable is quite unbalanced, which can cause problems in prediction  of urgency of Biopsy examination. It is important to note that recall is the most important of all parameters of the models, because we do not want to declare a patient with cancer as healthy.

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve

# Training Random Forest models with balanced classes
rf_model = RandomForestClassifier(random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest with Balanced Classes")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```


Although the accuracy of the model is good, the recall is quite poor

```python
# Support Vector Machine

from sklearn.svm import SVC

svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_test)

print("Support Vector Machine")
print(confusion_matrix(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

Recall only 0.45

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression


lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

# Probability prediction
y_pred_proba = lr_model.predict_proba(X_test)[:,1]

# Adjusting the threshold to maximize recall
threshold = 0.3 
y_pred_adjusted = (y_pred_proba >= threshold).astype(int)

# Model evaluation
print("Logistic Regression with Adjusted Threshold")
print(confusion_matrix(y_test, y_pred_adjusted))
print(classification_report(y_test, y_pred_adjusted))

# Calculation of ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred_proba)
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)

plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

With this approach, with the reduction of the threshold, we achieved a recall of 0.73
